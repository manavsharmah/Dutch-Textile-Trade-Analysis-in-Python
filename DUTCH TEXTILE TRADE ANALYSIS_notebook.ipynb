```python
#Dropping columns that are irrelevant
data=data.drop(['ship_captain','orig_date_src','orig_date','orig_exact_date_yn','orig_loc_port_arch','orig_loc_region_arch','dest_date_src','dest_date','dest_exact_date_yn','dest_loc_port_arch','dest_loc_region_arch','els_per_ps','additional_text_from_doc','notes_other'], axis=1)
data.head()
```

```python
data.info()
```

```python
# Define the column names to process
columns_to_process = ['total_value_guldens', 'total_value_stuivers', 'total_value_penningen', 'total_value_other', 'ps_value_gulden', 'ps_value_stuivers', 'ps_value_penningen']

# Remove commas from the string values
data[columns_to_process] = data[columns_to_process].replace(',', '', regex=True)

# Replace non-numeric values with 0 for the specified columns
data[columns_to_process] = data[columns_to_process].apply(pd.to_numeric, errors='coerce').fillna(0)


# Sum the columns and create a new column "total_value"
data['total_value'] = (
    data['total_value_guldens'].astype(float)
    + data['total_value_stuivers'].astype(float)
    + data['total_value_penningen'].astype(float)
    + data['total_value_other'].astype(float)
)

# Sum the columns and create a new column "price_per_unit"
data['price_per_unit'] = (    
    + data['ps_value_gulden'].astype(float)
    + data['ps_value_stuivers'].astype(float)
    + data['ps_value_penningen'].astype(float)
)

data.head(15)
```

```python
#Dropping columns that are irrelevant
data=data.drop(['total_value_guldens','total_value_stuivers','total_value_penningen','total_value_other','ps_value_gulden','ps_value_stuivers','ps_value_penningen','ps_value_currency'], axis=1)
data.head()
```

```python
# Define the columns to convert to integers
columns_to_convert = ['orig_yr', 'dest_yr']

# Replace non-finite values with a default value (e.g., -1)
data[columns_to_convert] = data[columns_to_convert].replace([np.nan, np.inf, -np.inf], -1)

# Convert selected columns to integer
data[columns_to_convert] = data[columns_to_convert].astype(int)
data.head(20)
```

```python
# Fill null values with 'NaN'
data = data.fillna('NaN')

# Save the DataFrame back to Excel
data.to_csv('Dutchtrade1.csv', index=False)
print("File saved successfully.")
```
